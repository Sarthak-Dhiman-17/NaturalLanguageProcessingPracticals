```python
import nltk
from collections import Counter
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
import numpy as np

nltk.download('punkt_tab')
```

```python
#text = "Sarthak, UID 22BAI70558, an AIML student at Chandigarh University, is passionate about AI and machine learning. He explores its applications through research and projects, aiming to contribute meaningfully to the field. With a strong academic foundation and proactive learning, he is a promising talent in AI/ML"
text = 'Hello! My name is Sarthak and I am a student of Chandigarh University, Currently I am persuing the degree of Artificial Intelligence and Machine Learning.'
```

```python
tokens = word_tokenize(text.lower())
print("Tokenized Words:", tokens)
word_freq = Counter(tokens)
print("\nWord Frequencies:", word_freq)
```

```python
def generate_ngrams(tokens, n):
    return list(ngrams(tokens, n))
# unigrams (n=1)
unigrams = generate_ngrams(tokens, 1)
print("\nunigrams:", unigrams)

# bigrams (n=2)
bigrams = generate_ngrams(tokens, 2)
print("\nBigrams:", bigrams)

# trigrams (n=3)
trigrams = generate_ngrams(tokens, 3)
print("\nTrigrams:", trigrams)

# multigrams (n=4)
multigrams = generate_ngrams(tokens, 4)
print("\nmultigrams:", multigrams)

multigrams = generate_ngrams(tokens, 5)
print("\nmultigrams:", multigrams)
```

```python
plt.figure(figsize=(10, 5))
plt.bar(word_freq.keys(), word_freq.values(), color='skyblue')
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Word Frequency in Text")
plt.xticks(rotation=45)
plt.show()
```

```python
# Generate WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(word_freq)

# Display WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("WordCloud of Text")
plt.show()
```

```python
# Generate rainbow colors for each bar
colors = plt.cm.rainbow(np.linspace(0, 1, len(word_freq)))

words, counts = zip(*word_freq.items())

# Plot the horizontal bar chart using Matplotlib
plt.figure(figsize=(10, 6))
plt.barh(words, counts, color=colors)  # Horizontal bars with rainbow colors

# Invert y-axis to match the style of the uploaded image
plt.gca().invert_yaxis()

# Add labels and title
plt.xlabel("Frequency", fontsize=12)
plt.ylabel("Words", fontsize=12)
plt.title("Word Frequency in Text", fontsize=14)
```
